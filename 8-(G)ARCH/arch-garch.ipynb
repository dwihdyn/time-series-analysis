# Predict random data volatility using ARCH & GARCH
<br> </br>

### Reason we use gauss random data is because normal data doesnt deviates much, hence forecasting volatility will be useless
<br> </br>

### Source : https://machinelearningmastery.com/develop-arch-and-garch-models-for-time-series-forecasting-in-python/

```
Steps:
1) Get high volatile data (White noise with increasing variance)
2) Plot ACF to check whether data change in variance are correlated overtime 
3) Find best p in AR(p) using PACF (not shown here)
4) Find best q in MA(q) using ACF (not shown here)
5) Build ARCH & GARCH using best p q
```

### Get volatile data : Here we use random data N(0, 0.01), and every new data, volatility multiplies by 0.01

````python
import random as rn
import matplotlib.pyplot as plt

# dataset 
rn.seed(1)   # random generator init | we need to make 1 random data
data = [rn.gauss(0, i*0.01) for i in range(1,100+1)]

# plot 
plt.plot(data)
plt.show()
````

### Plot ACF : to check whether data change in variance are correlated overtime. if corr > 0.2 up to lag 15, assumption pass

````python
from statsmodels.graphics.tsaplots import plot_acf
import numpy as np

# Convert data from list to array dataframe (pandas.DataFrame works too) 
squared_data = np.array([x**2 for x in data])

plot_acf(squared_data)
plt.show()
````

- Can see theres quite strong positive correlation of the variance up to lag15. Hence its reasonable satisfy ARCH assumption (variance must have correlation with their old own variance)
- From plot, p=15 for ARCH(15)

### Build ARCH(p) model

````python
from arch import arch_model

# split train & test data
n_test = 10
train, test = data[:-n_test], data[-n_test:]

# define model
model = arch_model(train, mean='Zero', vol='ARCH', p=15)

# fit data into model (run calculation/build model using training data)
model_fit = model.fit()

# Forecast data with same length as test data
yhat = model_fit.forecast(horizon=n_test)

# Plot actual vs forecasted data and see side by side
actual_var = [i*0.01 for i in range(0,100)]
plt.plot(actual_var[-n_test:], label='actual')
plt.plot(yhat.variance.values[-1,:], label='forecast')
plt.legend(loc='upper left', fontsize=8)
plt.show()
````

### Build GARCH(p,q)

````python
# define model
model = arch_model(train, mean='Zero', vol='GARCH', p=15, q=15)

# fit data into model (run calculation/build model using training data)
model_fit = model.fit()

# Forecast data with same length as test data
yhat = model_fit.forecast(horizon=n_test)

# Plot actual vs forecasted data and see side by side
actual_var = [i*0.01 for i in range(0,100)]
plt.plot(actual_var[-n_test:], label='actual')
plt.plot(yhat.variance.values[-1,:], label='forecast')
plt.legend(loc='upper left', fontsize=8)
plt.show()
````
