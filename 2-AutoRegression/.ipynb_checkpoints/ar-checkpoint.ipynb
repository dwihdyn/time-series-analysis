# Predicting Daily temperature using AutoRegression data
<br> </br>
Source :https://machinelearningmastery.com/autoregression-models-time-series-forecasting-python/

```
Steps :
1) Stationary the data using differencing
2) Check existance of correlation using lag_plot() or normal pearson corr(). if doesnt exist, go to PACF
3) Plot ACF to see which lags are to be used in prediction
4) Use those significant lags to AR for prediction. Use AR() package to auto-detect which lags are significant
```

````python
# see dataset & plot data

import pandas as pd
import matplotlib.pyplot as plt
series = pd.read_csv('daily-min-temperatures.csv', header=0, index_col=0)
print(series.head())

series.plot()
plt.show()
````

### Data seems stationary, before doing ACF, we can run a quick check whether autocorrelation do exist or not

# Quick check for autocorrelation

````python
# Method 1 : using scatter lag_plot
pd.plotting.lag_plot(series)
plt.show()

# Method 2 : use usual Pearson correlation (lag1 vs current)
values = pd.DataFrame(series.values)
df = pd.concat([values.shift(1), values], axis=1)
df.columns = ['t-1', 't+1']
result = df.corr()
print(result)
````

- Assumption of AR is observations at previous time steps are useful to predict the value at the next time step. Hence to be useful, past data must have some correlation with the current.

- Above plot we can see that there is positive correlation with the lagged data, hence assumption satisfied
- From Pearson correl, its highly correlated (0.77), hence assumption satisfied

# ACF

````python
# plot ACF using pandas
pd.plotting.autocorrelation_plot(series)
plt.show()

# using statsmodels
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(series, lags=35)
plt.show()
````

- Solid & Dashed line represent 95% & 99% confident interval for correlation values respectively | Outside dashed line -> Take that lag which shall be put into AR model
- Decays over time, hence we are dealing with Auto regressive process
- Take the lags that are above the line to be insert into AR model. OR use AR() that auto-detect those significant lags
#### We will 'predict' the most recent 7 data points, and compare it with the actual data to see how accurate our predition is

# AR Model prediction - FIXED

````python
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error   # lower MSE, accurate prediction

# split dataset. All dataset to build AR, last 7 data for comparation
X = series.values
train, test = X[1 : len(X)-7], X[len(X)-7 : ]

# build (train) AR model
model = AR(train)     # gives the best AR equation (cant be shown)
model_fit = model.fit()

# show how many lags parameter used & coefficients
print('Lag parameters : b0 + %s' % model_fit.k_ar)
print('Coefficients: %s' % model_fit.params)

# make predictions & measure accuracy using MSE
predictions = model_fit.predict(start = len(train), end = len(train) + len(test) - 1, dynamic = False)
print(f"predictions : {predictions}")
for i in range(len(predictions)):
    print('predicted=%f, actual=%f' % (predictions[i] , test[i]))
error = mean_squared_error(test, predictions)
print("Test MSE : %.3f" % error)

# plot results
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
````

# AR Model prediction - ROLLING

````python
# rolling : relearn AR as new data comes in

# split dataset
X = series.values
train, test = X[1:len(X)-7], X[len(X)-7:]


# build (train) AR model
model = AR(train)
model_fit = model.fit()
window = model_fit.k_ar
coef = model_fit.params


# walk forward over time steps in test
history = train[len(train)-window:]                   # get training data
history = [history[i] for i in range(len(history))]   # put trained data into array
predictions = list()                                  # empty list
for t in range(len(test)):
	length = len(history)
	lag = [history[i] for i in range(length-window,length)]
	yhat = coef[0]
	for d in range(window):
		yhat += coef[d+1] * lag[window-d-1]
	obs = test[t]
	predictions.append(yhat)
	history.append(obs)
	print('predicted=%f, expected=%f' % (yhat, obs))
error = mean_squared_error(test, predictions)
print('Test MSE: %.3f' % error)


# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
````

- Lower MSE, means forecasting are improved (read AR, MA, ARIMA in statsmodels.tsa.ar_model documentation)
