# ARIMA on some random data
<br> </br>
Source : https://www.machinelearningplus.com/time-series/arima-model-time-series-forecasting-python/

```
Steps :
1) Augmented Dickey-Fuller test to data to check whether data is stationary or not (whether differencing is needed or not)
2) Find best d in I(d) using ACF & ADF test
3) Find best p in AR(p) using PACF
4) Find best q in MA(q) using ACF
5) Build ARIMA(p,d,q) using 100% past data (and improve model depending on the ARIMA result & see any abnormal residual error)
6) Build ARIMA using 75% past data to forecast the 25% test data (secret to improve model is TRIAL & ERROR around the optimal (p,d,q) value until we get model that almost same with actual)
7) CONTINUE TO STEP 11: accuracy metrics
```

### Checking stationarity using ADF test

````python
import pandas as pd, numpy as np
from statsmodels.tsa.stattools import adfuller

df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/wwwusage.csv', names=['value'], header=0)
result = adfuller(df.value.dropna())
print(f"ADF Statistics : {result[0]}")
print(f"p-value : {result[1]}")

# p-value 0.12 (> 0.05), hence data is not stationary, differencing is needed
````

### Difference data (if needed) and observe its ACF to get optimal difference (strong or weak stationary)

````python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# chart size & settings
plt.rcParams.update({'figure.figsize':(12,7), 'figure.dpi':120})

# Original series
fig, axes = plt.subplots(3, 2, sharex=True)
axes[0, 0].plot(df.value)
axes[0, 0].set_title('Original Series')
plot_acf(df.value, ax=axes[0,1])

# 1st differencing
first = df.value.diff()
axes[1, 0].plot(first)
axes[1, 0].set_title('1st Order Differencing')
plot_acf(first.dropna(), ax=axes[1,1])

# 2nd differencing
second = df.value.diff().diff()
axes[2, 0].plot(second)
axes[2, 0].set_title('2nd Order Differencing')
plot_acf(second.dropna(), ax=axes[2,1])

plt.show()

ori = adfuller(df.value.dropna())
one = adfuller(first.dropna())
two = adfuller(second.dropna())

print('ADF P-Value for each differencing dataset')
print(f"Original : {ori[1]}")
print(f"One Differencing : {one[1]}")
print(f"Two Differencing : {two[1]}")
````

- From chart & ADF test, 2nd order is the one that reach stationary.
- But we can see in ACF for 2nd order, it goes from negative to positive correlation very quickly (at lag 3). Meaning the data is being OVER-differenced (will giva inaccurate result)
- Hence we will go with 1st order & assume weak stationary

````python
# Quick check on stationary 
# ndiffs : number of differences required for a stationary series
from pmdarima.arima.utils import ndiffs
y = df.value

# ADF
print(f"ADF : Need {ndiffs(y, test='adf')} differences for series to be stationary")

# PP
print(f"PP : Need {ndiffs(y, test='pp')} differences for series to be stationary")

# KPSS (not sure)
print(f"KPSS : {ndiffs(y, test='kpss')}")

print("ADF & PP shows that data is strong stationary if difference is 2")
````

### Find best p in AR(p)

````python
# chart size & settings
plt.rcParams.update({'figure.figsize' : (20,3),'figure.dpi' : 120})

# plot pacf using first differencing data
fig, axes = plt.subplots(1, 2, sharex = True)
axes[0].plot(first)
axes[0].set_title('1st Differencing')
axes[1].set(xlim=(0,20), ylim=(-5,5))
plot_pacf(first.dropna(), ax=axes[1])

plt.show()

# strangely, since 2nd order is over-difference & 1st order is under-difference, we using 1st order (under-difference), hence we shouldve set AR(p + 1) -> AR(2)
````

Despite the fact that 1 & 2 are outside significant limit (blue zone), to be conservative we shall take only 1. hence p = 1 -> AR(1)

### Find best q in MA(q)

````python
# chart size & settings
plt.rcParams.update({'figure.figsize' : (20,3),'figure.dpi' : 120})

# plot acf using first differencing data
fig, axes = plt.subplots(1, 2, sharex=True)
axes[0].plot(first)
axes[0].set_title('1st Differencing')
axes[1].set(xlim=(0,20), ylim=(-1.5,1.5))
plot_acf(first.dropna(), ax=axes[1])

plt.show()

print(first.tail())
````

- The lags that really stands out from significant limit is up tp 2. hence q=2 -> MA(2)

### Build ARIMA(p,d,q) using all data

````python
# from previos analysis, we concluded with ARIMA(1,1,2) 

from statsmodels.tsa.arima_model import ARIMA

# ARIMA(1,1,2)
model = ARIMA(df.value, order=(1,1,2))
model_fit = model.fit(disp=0)
print(model_fit.summary())
````

```
ARIMA(1,1,2) 
-> Yt = const_coef + ar.L1_coef (Yt-1) + ma.L1_coef (ERRt-1) + ma.L2_coef (ERRt-2)
-> Yt = 1.1202     + 0.6351 (-2)     + 0.5287 (ERRt-1)     + (-0.001) (ERRt-2)
```
<br> </br>

- From table, we can see that ma.L2_coef is almost 0 (almost gives no contribution) and the ma.L2_P>|z| (MA P-value) is highly insignificant (ideally < 0.05). 
- Hence removing MA2 could improve the model

````python
# try remove MA(2)

model = ARIMA(df.value, order=(1,1,1))
model_fit = model.fit(disp=0)
print(model_fit.summary())
````

- AIC reduced , p-value for ar.L1 & ma.L1 became highly significant (<< 0.05), hence ARIMA model has been improved

### Plot predicted residual error | to see whether any of the predicted value has massive error (if no. then the predicting model is a good fit)

````python
# plot residual error

residuals = pd.DataFrame(model_fit.resid)
fig, ax = plt.subplots(1,2)
residuals.plot(title = 'Residuals : Time Series', ax = ax[0])
residuals.plot(kind='kde', title = 'Residuals : Frequency/Density Curve', ax=ax[1])
plt.show()
````

- Can see on both time_series & freq_density chart that mean of error near 0 and near-consistent variance error. Hence model is a good fit

### Plot Actual vs Predicted data

````python
model_fit.plot_predict(dynamic=False)   # dynamic=False is to include lagged value (True will give straight line of trend)
plt.show()
````

### Build ARIMA with 75% data & test accuracy with 25% data

````python
# since we are dealing with time series data, sequence of data matters. hence cant use random sampling (must separate properly. here is 75:25)

# split data
train = df.value[ : 85]
test = df.value[85 : ]

# build model
model = ARIMA(train, order=(1,1,1))   # follow prev best setting
fitted = model.fit(disp=1)
print(fitted.summary())

# forecast data
forecast_data, series_data, confidence = fitted.forecast(15, alpha=0.05)   # forecast the 15 test data with 95% (conf)idence
fc_series = pd.Series(forecast_data, index=test.index)
lower_series = pd.Series(confidence[:,0], index=test.index)
upper_series = pd.Series(confidence[:,1], index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi = 100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='y', alpha=.15)
plt.title('Forecast vs Actual')
plt.legend(loc='upper left', fontsize=8)
plt.show()
````

- Actual & forecast data are within the confidence band (increase brightness to see band). Hence it fine
<br> </br>

- From chart, we can see all of the forecast data are below actual
- Hence to increase, we can increase (p,d,q) until its close to actual data, WHILE keeping AIC & p-value low (lower than prev model & < 0.05 respectively)

### Improve model by changing (p,d,q) so that it closest to actual test data, while keeping AIC & p-value low

````python
# Build model 
model = ARIMA(train, order=(3,2,1))   # play around here until forecase close to actual
fitted = model.fit(disp=1)
print(fitted.summary())

# forecast data
fc, se, conf = fitted.forecast(15, alpha = 0.05)   # 95% conf
fc_series = pd.Series(fc, index=test.index)
lower_series = pd.Series(conf[:,0], index=test.index)
upper_series = pd.Series(conf[:,1], index=test.index)

# Plot
plt.figure(figsize=(12,5), dpi=100)
plt.plot(train, label='training')
plt.plot(test, label='actual')
plt.plot(fc_series, label='forecast')
plt.fill_between(lower_series.index, lower_series, upper_series, color='y', alpha=.15)
plt.title('Forecast vs Actual')
plt.legend(loc='upper left', fontsize=8)
plt.show()
````

- AIC reduced, p-value for all ar & ma are < 0.05. Model do improved
- Key here to find best ARIMA : trial & error around 'optimal' p,d,q
